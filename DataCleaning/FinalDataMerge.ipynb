```python
import pandas as pd
```

## Dissolved Hg Dataset Merge

```python
# Import CSV file as Pandas Dataframes
GN01 = pd.read_csv('../Data/GN01/GN01_D.csv')
GP16 = pd.read_csv('../Data/GP16/GP16_HgD_cleaned.csv')
GT10_11 = pd.read_csv('../Data/GT10-11/GT10_11_HgD_cleaned.csv')
Met = pd.read_csv('../Data/Metzyme/Met_HgD_cleaned.csv')
```

```python
GN01.head()
```

```python
GN01.columns
```

```python
# Remove unneeded columns
GN01.drop(columns=['Unnamed: 0', 'Station_ID', 'Sample_Depth', 'Start_Latitude',
       'Start_Longitude', 'Hg0_pM', 'HgT_pM', 'DMHg_pM', 'TRANSM',
       'FLUORM', 'NITRIT'], inplace=True)
```

```python
GN01.columns
```

```python
# Update column names for merge
# GO-BGC Float Paramters: PRES TEMP PSAL DOXY CHLA BBP700 NITRATE
GN01.rename(columns = {'CTDPRS':'PRES', 'CTDTMP':'TEMP', 'CTDSAL':'PSAL', 'CTDOXY':'DOXY'}, inplace = True)
GN01.head()
```

```python
GN01.columns
```

```python
# Reorder columns
GN01 = GN01.reindex(columns=['MMHg_pM', 'PRES', 'TEMP', 'PSAL', 'DOXY', 'NITRAT','SILCAT','PHSPHT'])
GN01.head()
```

```python
GP16.head()
```

```python
GP16.columns
```

```python
# Remove unneeded columns
GP16.drop(columns=['STNNBR', 'depth', 'lat', 'lon', 'HgT_pM', 'Hg0_pM', 
       'DMHg_pM', 'NITRIT', 'TRANSM', 'FLUORM'], inplace=True)
GP16.columns
```

```python
# Update column names for merge
# GO-BGC Float Paramters: PRES TEMP PSAL DOXY CHLA BBP700 NITRATE
GP16.rename(columns = {'CTDPRS':'PRES', 'CTDTMP':'TEMP', 'CTDSAL':'PSAL', 'CTDOXY':'DOXY'}, inplace = True)
GP16.columns
```

```python
# Reorder columns
GP16 = GP16.reindex(columns=['MMHg_pM', 'PRES', 'TEMP', 'PSAL', 'DOXY', 'NITRAT','SILCAT','PHSPHT'])
GP16.head()
```

```python
GT10_11.head()
```

```python
# Convert MMHg concentrations from fM to pM
GT10_11['MMHg_pM'] = GT10_11['MMHg_fM']/1000
```

```python
GT10_11.columns
```

```python
# Remove unneeded columns
GT10_11.drop(columns=['Year', 'STNNBR', 'DEPTH', 'lat', 'lon', 'Hg_total', 'Hg0', 'MMHg_fM',
       'DMHg_fM', 'TRANSM', 'FLUORM',
       'NITRIT'], inplace=True)
GT10_11.columns
```

```python
# Update column names for merge
# GO-BGC Float Paramters: PRES TEMP PSAL DOXY CHLA BBP700 NITRATE
GT10_11.rename(columns = {'CTDPRS':'PRES', 'CTDTMP':'TEMP', 'CTDSAL':'PSAL', 'CTDOXY':'DOXY'}, inplace = True)
GT10_11.columns
```

```python
# Reorder columns
GT10_11 = GT10_11.reindex(columns=['MMHg_pM', 'PRES', 'TEMP', 'PSAL', 'DOXY', 'NITRAT','SILCAT','PHSPHT'])
GT10_11.head()
```

```python
Met.head()
```

```python
# Convert MMHg concentrations from fM to pM
Met['MMHg_pM'] = Met['MMHg_fM']/1000
```

```python
Met.columns
```

```python
# Remove unneeded columns
Met.drop(columns=['Station', 'Depth_m', 'long_E', 'lat_N', 
       'nitrite_umol_L', 'HgT_pM', 'Hg0_pM', 'DMHg_fM', 'MMHg_fM'], inplace=True)
Met.columns
```

```python
# Update column names for merge
# GO-BGC Float Paramters: PRES TEMP PSAL DOXY CHLA BBP700 NITRATE
Met.rename(columns = {'CTDPRS':'PRES', 'Temp_C':'TEMP', 'SALINITY_PSS78':'PSAL', 'Oxygen_umol_kg':'DOXY', 'PO4_umol_L':'PHSPHT','Silicate_umol_L':'SILCAT', 'nitrate_umol_L':'NITRAT'}, inplace = True)
Met.columns
```

```python
# Reorder columns
Met = Met.reindex(columns=['MMHg_pM', 'PRES', 'TEMP', 'PSAL', 'DOXY', 'NITRAT','SILCAT','PHSPHT'])
Met.head()
```

```python
# Concatenate 4 datasets 
Hg_D = pd.concat([GN01,GP16,GT10_11,Met])
Hg_D.head()
```

```python
# Remove unneeded columns
Hg_D.drop(columns=['SILCAT','PHSPHT'], inplace=True)
Hg_D.shape
```

```python
# Drop NaN entries
Hg_D = Hg_D.dropna()
Hg_D.shape
```

```python
# Remove depths >2000 meters
Hg_D = Hg_D.loc[Hg_D['PRES'] < 2000]
```

```python
# Export 
Hg_D.to_csv('../Data/Hg_D.csv')
```
