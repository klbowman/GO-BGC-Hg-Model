## GA10 Data Cleaning & Merging

```python
# Import CSV file as Pandas Dataframes
HgD = pd.read_csv('../Data/GA10/GA10_HgD.csv')
CTD = pd.read_csv('../Data/GA10/GA10_CTD.csv')
```

```python
# Merge HgD with Bottle
GA10_HgD = HgD.merge(CTD, how='left', on=['Station', 'depth'])
GA10_HgD.head()
```

```python
GA10_HgD.shape
```

```python
# Remove duplicate depths in GT16_HgD dataframe
GA10_HgD = GA10_HgD.groupby(by=['Station', 'depth']).mean()
GA10_HgD.shape
```

```python
GA10_HgD.head()
```

```python
GA10_HgD.columns
```

```python
# Remove data with bad quality codes
# https://exchange-format.readthedocs.io/en/latest/quality.html

# CTD codes (keep codes 2 & 3)

GA10_HgD = GA10_HgD.loc[(GA10_HgD['CTDSAL_FLAG_W'] > 1) | (GA10_HgD['CTDSAL_FLAG_W'] < 4)]
GA10_HgD = GA10_HgD.loc[(GA10_HgD['CTDOXY_FLAG_W'] > 1) | (GA10_HgD['CTDOXY_FLAG_W'] < 4)]
GA10_HgD = GA10_HgD.loc[(GA10_HgD['SILCAT_FLAG_W'] > 1) | (GA10_HgD['SILCAT_FLAG_W'] < 4)]


# Water quality codes (remove codes 1, 4, 5, 9)
GA10_HgD = GA10_HgD.loc[(GA10_HgD['NITRAT_FLAG_W'] > 1) | (GA10_HgD['NITRAT_FLAG_W'] < 4)| (GA10_HgD['NITRAT_FLAG_W'] == 6)]
GA10_HgD = GA10_HgD.loc[(GA10_HgD['NITRIT_FLAG_W'] > 1) | (GA10_HgD['NITRIT_FLAG_W'] < 4)| (GA10_HgD['NITRIT_FLAG_W'] == 6)]
GA10_HgD = GA10_HgD.loc[(GA10_HgD['PHSPHT_FLAG_W'] > 1) | (GA10_HgD['PHSPHT_FLAG_W'] < 4)| (GA10_HgD['PHSPHT_FLAG_W'] == 6)]

GA10_HgD.shape
```

```python
# Remove unneeded columns

GA10_HgD.drop(columns=['CTDPRS_FLAG_W',  'CTDTMP_FLAG_W', 
       'CTDSAL_FLAG_W','CTDOXY_FLAG_W',  'PHSPHT_FLAG_W',
        'SILCAT_FLAG_W', 'NITRIT_FLAG_W', 'NITRAT_FLAG_W',
      ], inplace=True)

GA10_HgD.head()
```

```python
# Export 
GA10_HgD.to_csv('../Data/GA10/GA10_HgD_cleaned.csv')
```
