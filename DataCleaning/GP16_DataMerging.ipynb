```python
import pandas as pd
```

## GP16 Data Cleaning & Merging

```python
# Import CSV file as Pandas Dataframes
HgD = pd.read_csv('../Data/GP16/GP16_HgD.csv')
Bottle = pd.read_csv('../Data/GP16/GP16_Bottle.csv')
Profile = pd.read_csv('../Data/GP16/GP16_Profile.csv')
```

```python
HgD.head()
```

```python
Bottle.head()
```

```python
HgD.shape
```

```python
# Merge HgD with Bottle
GT16_HgD = HgD.merge(Bottle, how='left', on=['STNNBR', 'depth'])
GT16_HgD.head()
```

```python
GT16_HgD.shape
```

```python
# Remove duplicate depths in GT16_HgD dataframe
GT16_HgD = GT16_HgD.groupby(by=['STNNBR', 'depth']).mean()
GT16_HgD.shape
```

```python
Profile.head()
```

```python
# Merge GT16_HgD with Profile
GT16_HgD = GT16_HgD.merge(Profile, how='left', on=['STNNBR', 'depth'])
GT16_HgD.head()
```

```python
GT16_HgD.shape
```

```python
# Remove duplicate depths in GT16_HgD dataframe
GT16_HgD = GT16_HgD.groupby(by=['STNNBR', 'depth']).mean()
GT16_HgD.shape
```

```python
GT16_HgD.columns
```

```python
# Remove data with bad quality codes
# https://exchange-format.readthedocs.io/en/latest/quality.html

# CTD codes (keep codes 2 & 3)

GT16_HgD = GT16_HgD.loc[(GT16_HgD['CTDSAL_FLAG_W'] > 1) | (GT16_HgD['CTDSAL_FLAG_W'] < 4)]
GT16_HgD = GT16_HgD.loc[(GT16_HgD['CTDOXY_FLAG_W'] > 1) | (GT16_HgD['CTDOXY_FLAG_W'] < 4)]
GT16_HgD = GT16_HgD.loc[(GT16_HgD['SILCAT_FLAG_W'] > 1) | (GT16_HgD['SILCAT_FLAG_W'] < 4)]
GT16_HgD = GT16_HgD.loc[(GT16_HgD['TRANSM_FLAG_W'] > 1) | (GT16_HgD['TRANSM_FLAG_W'] < 4)]
GT16_HgD = GT16_HgD.loc[(GT16_HgD['FLUORM_FLAG_W'] > 1) | (GT16_HgD['FLUORM_FLAG_W'] < 4)]

# Water quality codes (remove codes 1, 4, 5, 9)


GT16_HgD = GT16_HgD.loc[(GT16_HgD['HgT_flag'] > 1) | (GT16_HgD['HgT_flag'] < 4)| (GT16_HgD['HgT_flag'] == 6)]
GT16_HgD = GT16_HgD.loc[(GT16_HgD['Hg0_flag'] > 1) | (GT16_HgD['Hg0_flag'] < 4)| (GT16_HgD['Hg0_flag'] == 6)]
GT16_HgD = GT16_HgD.loc[(GT16_HgD['MMHg_flag'] > 1) | (GT16_HgD['MMHg_flag'] < 4)| (GT16_HgD['MMHg_flag'] == 6)]
GT16_HgD = GT16_HgD.loc[(GT16_HgD['DMHg_flag'] > 1) | (GT16_HgD['DMHg_flag'] < 4)| (GT16_HgD['DMHg_flag'] == 6)]

GT16_HgD = GT16_HgD.loc[(GT16_HgD['NITRAT_FLAG_W'] > 1) | (GT16_HgD['NITRAT_FLAG_W'] < 4)| (GT16_HgD['NITRAT_FLAG_W'] == 6)]
GT16_HgD = GT16_HgD.loc[(GT16_HgD['NITRIT_FLAG_W'] > 1) | (GT16_HgD['NITRIT_FLAG_W'] < 4)| (GT16_HgD['NITRIT_FLAG_W'] == 6)]
GT16_HgD = GT16_HgD.loc[(GT16_HgD['PHSPHT_FLAG_W'] > 1) | (GT16_HgD['PHSPHT_FLAG_W'] < 4)| (GT16_HgD['PHSPHT_FLAG_W'] == 6)]

GT16_HgD.shape
```

```python
# Remove unneeded columns

GT16_HgD.drop(columns=['HgT_flag','Hg0_flag','MMHg_flag', 'DMHg_flag', 'LATITUDE', 'LONGITUDE',
       'GEOTRC_SAMPNO_y','SILCAT_FLAG_W','NITRAT_FLAG_W',
       'NITRIT_FLAG_W',  'PHSPHT_FLAG_W',
       'CTDPRS_FLAG_W', 'CTDTMP_FLAG_W', 'CTDSAL_FLAG_W',
       'CTDOXY_FLAG_W',  'TRANSM_FLAG_W', 
       'FLUORM_FLAG_W'], inplace=True)

GT16_HgD.head()
```

```python
# Export 
GT16_HgD.to_csv('../Data/GP16/GP16_HgD_cleaned.csv')
```

```python
# Import GP16_HgP_merged file as Pandas Dataframe
    # Particulate Hg, POC and GN01_D file merged in SQL with depths matching +/- 5 meters
GP16_HgP_merged = pd.read_csv('../Data/GP16/GP16_HgP_merged.csv')
GP16_HgP_merged.head()
```

```python
GP16_HgP_merged.shape
```

```python
# Remove duplicate depths in GN01_P dataframe
GP16_HgP_merged = GP16_HgP_merged.groupby(by=['STNNBR', 'depth']).mean()
GP16_HgP_merged.shape
```

```python
GP16_HgP_merged.columns
```

```python
# Remove data with bad quality codes
# https://exchange-format.readthedocs.io/en/latest/quality.html

# Water quality codes (remove codes 1, 4, 5, 9)

GP16_HgP_merged = GP16_HgP_merged.loc[(GP16_HgP_merged['MMHgP_flag'] > 1) | (GP16_HgP_merged['MMHgP_flag'] < 4)| (GP16_HgP_merged['MMHgP_flag'] == 6)]
GP16_HgP_merged = GP16_HgP_merged.loc[(GP16_HgP_merged['HgP_flag'] > 1) | (GP16_HgP_merged['HgP_flag'] < 4)| (GP16_HgP_merged['HgP_flag'] == 6)]
GP16_HgP_merged = GP16_HgP_merged.loc[(GP16_HgP_merged['POC_sink_flag'] > 1) | (GP16_HgP_merged['POC_sink_flag'] < 4)| (GP16_HgP_merged['POC_sink_flag'] == 6)]
GP16_HgP_merged = GP16_HgP_merged.loc[(GP16_HgP_merged['POC_susp_flag'] > 1) | (GP16_HgP_merged['POC_susp_flag'] < 4)| (GP16_HgP_merged['POC_susp_flag'] == 6)]

GP16_HgP_merged.shape
```

```python
# Remove unneeded columns

GP16_HgP_merged.drop(columns=['HgP_flag', 
       'MMHgP_flag', 'POC_sink_flag', 'POC_susp_flag'], inplace=True)

GP16_HgP_merged.head()
```

```python
# Export 
GP16_HgP_merged.to_csv('../Data/GP16/GP16_HgP_cleaned.csv')
```
