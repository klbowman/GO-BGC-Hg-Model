```python
import pandas as pd
```

## GT10-11 Data Cleaning & Merging

```python
# Import CSV file as Pandas Dataframes
HgD = pd.read_csv('../Data/GT10-11/GT10_11_HgD.csv')
HgP = pd.read_csv('../Data/GT10-11/GT10_11_HgP.csv')
GTC_10 = pd.read_csv('../Data/GT10-11/GT10_Profiles.csv')
GTC_11 = pd.read_csv('../Data/GT10-11/GT11_Profiles.csv')
Bottle_10 = pd.read_csv('../Data/GT10-11/GT10_Bottle.csv')
Bottle_11 = pd.read_csv('../Data/GT10-11/GT11_Bottle.csv')
```

```python
HgD.head()
```

```python
# Update column names for merge
HgD.rename(columns = {'sta':'STNNBR', 'depth':'DEPTH'}, inplace = True)
HgD.head()
```

```python
HgD.shape
```

```python
HgD.dtypes
```

```python
GTC_10.shape
```

```python
GTC_11.shape
```

```python
# Concatenate GTC 2010 & 2011
GTC = pd.concat([GTC_10, GTC_11])
GTC.head()
```

```python
GTC.shape
```

```python
# Merge HgD with GTC
GT10_11_D = HgD.merge(GTC, how='left', on=['Year', 'STNNBR','DEPTH'])
GT10_11_D.head()
```

```python
GT10_11_D.shape
```

```python
# Remove duplicate depths in GT10_11_D dataframe
GT10_11_D = GT10_11_D.groupby(by=['Year', 'STNNBR','DEPTH']).mean()
GT10_11_D.shape
```

```python
# Check bottle dataframe sizes
Bottle_10.shape
```

```python
# Check bottle dataframe sizes
Bottle_11.shape
```

```python
# Concatenate Bottle10 & Bottle11
Bottle = pd.concat([Bottle_10, Bottle_11])
Bottle.head()
```

```python
Bottle.shape
```

```python
# Update column names for merge
Bottle.rename(columns = {'CTDDEPTH':'DEPTH'}, inplace = True)
Bottle.head()
```

```python
# Merge GT10_11_D with Bottle
GT10_11_D = GT10_11_D.merge(Bottle, how='left', on=['Year', 'STNNBR','DEPTH'])
GT10_11_D.head()
```

```python
GT10_11_D.shape   
```

```python
GT10_11_D.columns
```

```python
# Remove data with bad quality codes
# https://exchange-format.readthedocs.io/en/latest/quality.html

# CTD codes (keep codes 2 & 3)

GT10_11_D = GT10_11_D.loc[(GT10_11_D['CTDSAL_FLAG_W'] > 1) | (GT10_11_D['CTDSAL_FLAG_W'] < 4)]
GT10_11_D = GT10_11_D.loc[(GT10_11_D['CTDOXY_FLAG_W'] > 1) | (GT10_11_D['CTDOXY_FLAG_W'] < 4)]
GT10_11_D = GT10_11_D.loc[(GT10_11_D['SILCAT_FLAG_W'] > 1) | (GT10_11_D['SILCAT_FLAG_W'] < 4)]
GT10_11_D = GT10_11_D.loc[(GT10_11_D['TRANSM_FLAG_W'] > 1) | (GT10_11_D['TRANSM_FLAG_W'] < 4)]
GT10_11_D = GT10_11_D.loc[(GT10_11_D['FLUORM_FLAG_W'] > 1) | (GT10_11_D['FLUORM_FLAG_W'] < 4)]
GT10_11_D = GT10_11_D.loc[(GT10_11_D['DEPTH_FLAG_W'] > 1) | (GT10_11_D['DEPTH_FLAG_W'] < 4)]

# Water quality codes (remove codes 1, 4, 5, 9)


GT10_11_D = GT10_11_D.loc[(GT10_11_D['Hg_total_flag'] > 1) | (GT10_11_D['Hg_total_flag'] < 4)| (GT10_11_D['Hg_total_flag'] == 6)]
GT10_11_D = GT10_11_D.loc[(GT10_11_D['Hg0_flag'] > 1) | (GT10_11_D['Hg0_flag'] < 4)| (GT10_11_D['Hg0_flag'] == 6)]
GT10_11_D = GT10_11_D.loc[(GT10_11_D['MMHg_flag'] > 1) | (GT10_11_D['MMHg_flag'] < 4)| (GT10_11_D['MMHg_flag'] == 6)]
GT10_11_D = GT10_11_D.loc[(GT10_11_D['DMHg_flag'] > 1) | (GT10_11_D['DMHg_flag'] < 4)| (GT10_11_D['DMHg_flag'] == 6)]

GT10_11_D = GT10_11_D.loc[(GT10_11_D['NITRAT_FLAG_W'] > 1) | (GT10_11_D['NITRAT_FLAG_W'] < 4)| (GT10_11_D['NITRAT_FLAG_W'] == 6)]
GT10_11_D = GT10_11_D.loc[(GT10_11_D['NITRIT_FLAG_W'] > 1) | (GT10_11_D['NITRIT_FLAG_W'] < 4)| (GT10_11_D['NITRIT_FLAG_W'] == 6)]
GT10_11_D = GT10_11_D.loc[(GT10_11_D['PHSPHT_FLAG_W'] > 1) | (GT10_11_D['PHSPHT_FLAG_W'] < 4)| (GT10_11_D['PHSPHT_FLAG_W'] == 6)]

GT10_11_D.head()
GT10_11_D.shape
```

```python
GT10_11_D.columns
```

```python
# Remove unneeded columns

GT10_11_D.drop(columns=['Hg_total_flag',
        'Hg0_flag', 'MMHg_flag', 'DMHg_flag',
       'LATITUDE_x', 'LONGITUDE_x','CTDPRS_FLAG_W', 
       'CTDTMP_FLAG_W',  'CTDSAL_FLAG_W',  'CTDOXY_FLAG_W',
       'TRANSM_FLAG_W',  'FLUORM_FLAG_W', 'DEPTH_FLAG_W',
       'LATITUDE_y', 'LONGITUDE_y',  'SILCAT_FLAG_W', 
       'NITRAT_FLAG_W',  'NITRIT_FLAG_W', 'PHSPHT_FLAG_W'], inplace=True)

GT10_11_D.head()
```

```python
# Export 
GT10_11_D.to_csv('../Data/GT10-11/GT10_11_HgD_cleaned.csv')
```

```python
# Import GN01_P CSV file as Pandas Dataframe
    # Particulate Hg, POC and GN01_D file merged in SQL with depths matching +/- 5 meters
GT10_11_HgP_merged = pd.read_csv('../Data/GT10-11/GT10_11_HgP_merged.csv')
GT10_11_HgP_merged.head()
```

```python
GT10_11_HgP_merged.shape
```

```python
# Remove duplicate depths in GN01_P dataframe
GT10_11_HgP_merged = GT10_11_HgP_merged.groupby(by=['Year','STNNBR', 'DEPTH']).mean()
GT10_11_HgP_merged.shape
```

```python
GT10_11_HgP_merged.columns
```

```python
# Remove data with bad quality codes
# https://exchange-format.readthedocs.io/en/latest/quality.html

# Water quality codes (remove codes 1, 4, 5, 9)

GT10_11_HgP_merged = GT10_11_HgP_merged.loc[(GT10_11_HgP_merged['MMHgP_flag'] > 1) | (GT10_11_HgP_merged['MMHgP_flag'] < 4)| (GT10_11_HgP_merged['MMHgP_flag'] == 6)]
GT10_11_HgP_merged = GT10_11_HgP_merged.loc[(GT10_11_HgP_merged['HgP_flag'] > 1) | (GT10_11_HgP_merged['HgP_flag'] < 4)| (GT10_11_HgP_merged['HgP_flag'] == 6)]
GT10_11_HgP_merged = GT10_11_HgP_merged.loc[(GT10_11_HgP_merged['POC_sink_flag'] > 1) | (GT10_11_HgP_merged['POC_sink_flag'] < 4)| (GT10_11_HgP_merged['POC_sink_flag'] == 6)]
GT10_11_HgP_merged = GT10_11_HgP_merged.loc[(GT10_11_HgP_merged['POC_susp_flag'] > 1) | (GT10_11_HgP_merged['POC_susp_flag'] < 4)| (GT10_11_HgP_merged['POC_susp_flag'] == 6)]

GT10_11_HgP_merged.shape
```

```python
# Remove unneeded columns

GT10_11_HgP_merged.drop(columns=['HgP_flag', 
       'MMHgP_flag', 'POC_sink_flag', 'POC_susp_flag'], inplace=True)

GT10_11_HgP_merged.head()
```

```python
# Export 
GT10_11_HgP_merged.to_csv('../Data/GT10-11/GT10_11_HgP_cleaned.csv')
```
